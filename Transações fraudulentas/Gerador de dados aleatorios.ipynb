```python
from faker import Faker
import pandas as pd
import numpy as np
```

```python
#Configuração
fake = Faker()
np.random.seed(42)
```

```python
# Parâmetros do banco de dados

n_records = 1000
```

```python
# Função para criar transações fictícias

def create_transaction():
    return {
        "IDTransação": fake.uuid4(),
        "IDpessoal": fake.uuid4(),
        "DatadaTransação": fake.date_time_this_year(),
        "Valor": round(np.random.uniform(10,1000), 2),
        "Empresa": fake.company(),
        "Categoria": np.random.choice(["Padaria", "Eletronicos", "Roupas", "Viagens", "Mercado", "Entretenimento"]),
        "Fraude": np.random.choice([0,1], p=[0.97 , 0.03]) 
    }
```

```python
# DataFrame
data = [create_transaction() for i in range(n_records)]
df = pd.DataFrame(data)
```
